```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
```

```python
# Assign the iris dataset features to the variable X
X = iris.data

# Assign the iris dataset target to the variable y
y = iris.target

# Store the species names in the species_names variable
species_names = iris.target_names

# Store the feature names in the feature_names variable
feature_names = iris.feature_names

# Loop through each of the four features
for i in range(len(feature_names)):
    # Create a new figure for each feature
    plt.figure()

    # Plot a boxplot of the feature values for each species
    data = []
    for j in range(3):
        feature = X[y == j, i]
        data.append(feature)
    plt.boxplot(data)

    # Add x-axis labels for each species
    plt.xticks([1, 2, 3], species_names)

    # Add the x-axis label "Species"
    plt.xlabel('Species')

    # Add the y-axis label with the feature name
    plt.ylabel(feature_names[i])

    # Add a title to the plot with the feature name and "Boxplot by Species"
    plt.title(f'Boxplot of {feature_names[i]} by Species')

# Show the plot
plt.show()
```

```python
for i in [2, 0]:
    plt.figure()
    for j in range(3):
        plt.scatter(X[y == j, i], X[y == j, i + 1], label=species_names[j])
    plt.xlabel('Length')
    plt.ylabel('Width')
    plt.title(f'Scatter plot of {["petal", "sepal"][i // 2]} length vs width')
    plt.legend()
    
plt.show()
```

```python
print('Petal length < 2.5 cm and petal width < 0.75 cm: Iris Setosa')
print('Petal length >= 2.5 cm and petal width >= 1.0 cm: Iris Virginica')
print('Petal length >= 2.5 cm and petal width < 1.0 cm: Iris Versicolour')
```
